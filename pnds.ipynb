# Filtro

```python
df[df.Rating > 8].head(2)
```

```python
df.Rating.value_counts()
```

```python
counts = df.Rating.value_counts()

counts.index
```

```python
counts.values
```

```python
counts_df = pd.DataFrame({'calif': counts.index, 'frequency': counts.values})
counts_df.head(4)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(
    data=counts_df,
    x='calif',
    y='frequency'
)

plt.title('Frecuencia de Calificaciones')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
# kde=True añade una curva de densidad suave sobre las barras
sns.histplot(data=df, x='Rating', kde=True, bins=25, color='skyblue')

plt.title('Distribución de las Calificaciones de Películas (Rating)')
plt.xlabel('Calificación')
plt.ylabel('Cantidad de Películas')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
# Graficamos Rating en el eje X y Revenue_Millions en el eje Y
sns.scatterplot(data=df, x='Rating', y='Revenue_Millions', alpha=0.6, color='coral', edgecolor='black')

plt.title('Relación entre Calificación e Ingresos (Revenue)')
plt.xlabel('Calificación (Rating)')
plt.ylabel('Ingresos en Millones (Revenue Millions)')
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='Runtime (Minutes)', fill=True, color='darkviolet', alpha=0.4)

plt.title('Distribución Suavizada de la Duración de las Películas')
plt.xlabel('Duración en Minutos (Runtime)')
plt.ylabel('Densidad de Probabilidad')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
# Filtramos los últimos 3 años para comparar su distribución
df_reciente = df[df['Year'].isin([2014, 2015, 2016])]

sns.violinplot(data=df_reciente, x='Year', y='Runtime (Minutes)', palette='pastel')

plt.title('Distribución de Duración de Películas en Años Recientes')
plt.xlabel('Año')
plt.ylabel('Duración en Minutos')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
# Seleccionamos únicamente variables numéricas relevantes
columnas_numericas = df[['Rating', 'Votes', 'Revenue_Millions', 'Metascore', 'Runtime (Minutes)']]
matriz_corr = columnas_numericas.corr()

# annot=True muestra los números dentro de los cuadros
sns.heatmap(matriz_corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

plt.title('Mapa de Calor de Correlaciones Numéricas')
plt.show()
```
